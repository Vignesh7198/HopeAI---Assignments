```python
import pandas as pd
import numpy as np
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
x = dataset.iloc[:,3:]
x
```

```python
from sklearn.cluster import OPTICS
```

```python
optics = OPTICS(min_samples=5, max_eps=np.inf, metric='minkowski', p=2, metric_params=None, cluster_method='xi', eps=None, xi=0.10, predecessor_correction=True, min_cluster_size=None, algorithm='auto', leaf_size=30, n_jobs=-1)
```

```python
cluster = optics.fit_predict(x)
```

```python
unsupervised = dataset.copy()
```

```python
unsupervised["cluster"] = cluster
```

```python
unsupervised
```

```python
import seaborn as sb
sb.lmplot(data=unsupervised,x=unsupervised.columns[3],y=unsupervised.columns[4],hue=unsupervised.columns[5], fit_reg=False)
```
