```python
import pandas as pd
import numpy as np
from sklearn.cluster import AffinityPropagation
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x = dataset.iloc[:,3:].values
```

```python
x
```

```python
cluster = AffinityPropagation(damping=0.5,preference=-15000,max_iter=200,convergence_iter=15,affinity="euclidean")
```

```python
label = cluster.fit_predict(x)
```

```python
label
```

```python
unsupervised = dataset.copy()
```

```python
unsupervised["Cluster"] = label
```

```python
unsupervised
```

```python
import seaborn as sns
sns.lmplot(data=unsupervised,x=unsupervised.columns[3],y=unsupervised.columns[4],hue=unsupervised.columns[5],fit_reg=False)
```
