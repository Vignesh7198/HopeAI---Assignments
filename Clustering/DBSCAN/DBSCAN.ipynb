```python
import pandas as pd
dataset = pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x = dataset.iloc[:,3:]
```

```python
from sklearn.cluster import DBSCAN
```

```python
cluster = DBSCAN(eps=10,min_samples=5,metric='euclidean', metric_params=None,leaf_size=30,p=2,n_jobs=1)
```

```python
predictedClusters = cluster.fit_predict(x)
```

```python
unsupervised = dataset.copy()
unsupervised["cluster"] = predictedClusters
```

```python
import seaborn as sb 
```

```python
sb.lmplot(data=unsupervised, x=unsupervised.columns[3],y=unsupervised.columns[4],hue=unsupervised.columns[5],fit_reg=False)
```
