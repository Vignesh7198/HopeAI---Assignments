```python
import pandas as pd
dataset = pd.read_csv("PrePlacement.csv")

dataset
```

```python
def stdNBgraph(dataset):
     from sklearn.preprocessing import StandardScaler
     sc = StandardScaler()
     sc.fit(dataset)
     return sc.transform(dataset)
```

```python
SNB = stdNBgraph(dataset['salary'].to_frame())
SNB
```

```python
import seaborn as sns
sns.distplot(SNB,kde=True)
```

```python
import numpy as np
newSal = np.log(SNB + 1)
```

```python
sns.distplot(newSal,kde=True)
```
