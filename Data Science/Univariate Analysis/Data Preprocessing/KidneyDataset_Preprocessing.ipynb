```python
    def quanQual(dataset):
        quan = []
        qual = []
        for data in dataset.columns:
            if(dataset[data].dtype=='object'):
                qual.append(data)
            else:
                quan.append(data)
        return quan,qual
```

```python
quan,qual = quanQual(dataset)
```

```python
dataset[quan]
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan,strategy='mean')
imp2 = SimpleImputer(strategy='most_frequent')
imp.fit(dataset[quan])
imp2.fit(dataset[qual])
quandataset = imp.transform(dataset[quan])
qualdataset = imp2.transform(dataset[qual])
```

```python
quandataset = pd.DataFrame(quandataset,columns=[quan])
qualdataset = pd.DataFrame(qualdataset,columns=[qual])
quandataset.info()
quandataset
```

```python
dataset = pd.concat([quandataset,qualdataset],axis=1)
```
