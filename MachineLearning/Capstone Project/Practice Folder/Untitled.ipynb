```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("dataset_traffic_accident_prediction1.csv")
dataset
```

```python
dataset[dataset.columns].corr()
```

```python
dataset.info()
```

```python
category = dataset[dataset.columns[dataset.dtypes=="object"]]
category
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
categorytransformed = si.fit_transform(category)

cat = pd.DataFrame(categorytransformed, columns=category.columns)
```

```python
df = dataset.copy()
num = df.drop(cat.columns,axis=1)
num
```

```python
num.isnull().sum()
```

```python
num.describe()
```

```python
num["Speed_Limit"].fillna(num["Speed_Limit"].median(),inplace=True)
num["Traffic_Density"].fillna(num["Traffic_Density"].mode()[0],inplace=True)
num["Number_of_Vehicles"].fillna(num["Number_of_Vehicles"].mode()[0], inplace=True)
num["Driver_Alcohol"].fillna(num["Driver_Alcohol"].mode()[0], inplace=True)
num["Driver_Age"].fillna(num["Driver_Age"].mode()[0], inplace=True)
num["Driver_Experience"].fillna(num["Driver_Experience"].mode()[0], inplace=True)
num["Accident"].fillna(num["Accident"].mode()[0], inplace=True)
```

```python
num.isnull().sum()
```

```python
df = pd.concat([cat,num],axis=1)
```

```python
df = pd.get_dummies(df,dtype=int,drop_first=True)
x = df.drop("Accident",axis=1)
y= df["Accident"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=10,criterion="entropy",random_state=0)
random_forest.fit(x_train,y_train)
random_forest.feature_importances_,x_train.columns
```

```python
pd.Series(random_forest.feature_importances_,index=x_train.columns).sort_values(ascending=False)
```
