## Load the dataset (Anxiety.csv)

```python
import pandas as pd
dataset = pd.read_csv("Anxiety.csv")
dataset
```

## Checking the presence of null value

```python
dataset.isnull().sum()
```

# Dataset appears to be good, without any null values

# Let's remove the unwanted columns 

```python
x = dataset.iloc[:,0:7]
```

```python
x.head(2)
```

```python
y=dataset.iloc[:,-1]
```

```python
y.value_counts()
```

## This seems to be a multiclass statement, let's convert this to a binary class, coz accuracy will drop when there is more classes and less number of samples

```python
y = y.map(lambda y: "Less Anxious" if y=="Minimal Anxiety" else "More Anxious");
```

```python
y.value_counts()
```

## Lets's join the dependent (y) and independent (x) variable and save it in csv file for future process

```python
dataset = pd.concat([x,y],axis=1)
```
