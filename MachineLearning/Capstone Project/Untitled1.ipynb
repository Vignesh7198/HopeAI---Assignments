```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("Anxiety.csv")
```

```python
dataset
```

```python
x = dataset.iloc[:,0:7]
y = dataset.iloc[:,-1]
```

```python
y = y.map(lambda y: "Less Anxious" if y=="Minimal Anxiety" else "More Anxious");
```

```python
x = pd.get_dummies(x, drop_first=True,dtype=int)
y= pd.get_dummies(y,drop_first=True,dtype=int)
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=10,criterion="gini",random_state=0)
random_forest.fit(x_train,y_train)
```

```python
pd.Series(random_forest.feature_importances_,index=x_train.columns).sort_values(ascending=False)
```

```python
y_predict = random_forest.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test,y_predict)
print(report)
```

```python

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

estimator = {
    'criterion':['gini', 'entropy', 'log_loss'],
    'max_features':['sqrt', 'log2', None],
    'n_estimators':[50,100]

}

grid = GridSearchCV(RandomForestClassifier(),estimator,verbose=3,refit=True,n_jobs=-1)
```

```python
re = grid.fit(x,y)
```

```python
re
```

```python
table = pd.DataFrame.from_dict(re.cv_results_)
```
