```python
import pandas as pd
dataset = pd.get_dummies(pd.read_csv('insurance_pre.csv'),drop_first=True,dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age','bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(independent,dependent,test_size=0.3,random_state=0);
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
paramgrid = {'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
             'C':[10,100,500,1000,2000,3000]}

grid = GridSearchCV(SVR(),paramgrid,refit=True,verbose = 3, n_jobs=-1)
re = grid.fit(x_train,y_train)
```

```python
re.cv_results_
```

```python
table = pd.DataFrame.from_dict(re.cv_results_)
```

```python
table
```

```python
print(f'Best Score:',grid.best_params_)
```
