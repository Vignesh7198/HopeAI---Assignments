```python
import pandas as pd
import numpy as np
```

```python
i_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
items = pd.read_csv('ml-100k/u.item', sep='|', names=i_cols,encoding='latin-1')
```

```python
items.head()
```

```python
genre = items.iloc[:,6:]
```

```python
movie_id_title = items.iloc[:,:2]
```

```python
movie_content=pd.concat([movie_id_title,genre],axis=1)
movie_content
```

```python
movie_content=movie_content.T
movie_content
```

```python
ratings = pd.read_csv('user_based/ml-100k/u.data',sep='\t',names=['user_id','movie_id','rating','unix_timestamp'])
```

```python
ratings
```

```python
movie_id_rating = ratings.groupby('movie_id')["rating"].mean().sort_values(ascending=False)
```

```python
movie_id_rating
```

```python
movie_content=movie_content.T
```

```python
movie_content['popularity'] = movie_content['movie id'].map(movie_id_rating)
```

```python
movie_content
```

```python
movie_content[(movie_content['Action']==1)].sort_values(by='popularity', ascending=False).head(5)['movie id'].values
```

```python
popular_genre=[]
for i in movie_content.columns[2:20]:
    popular_genre.append(movie_content[(movie_content[i]==1)].sort_values(by='popularity', ascending=False).head(5)['movie title'].values)
```

```python
pd.DataFrame(popular_genre,index=movie_content.columns[2:20], columns=[1,2,3,4,5])
```
