```python
import pandas as pd
dataset = pd.get_dummies(pd.read_csv("CKD.csv"), drop_first=True,dtype=int)
pd.set_option('display.max_columns', None)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent = dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
x_train, x_test, y_train, y_test = train_test_split(independent,dependent,test_size=0.3,random_state=0)
from sklearn.linear_model import LogisticRegression
params={
         "solver":["lbfgs", "liblinear", "newton-cg", "newton-cholesky", "sag", "saga"],
          "penalty":["l2"]
}
grid = GridSearchCV(LogisticRegression(),params,verbose=1,n_jobs=-1,refit=True, scoring="f1_weighted")
```

```python
grid.fit(x_train,y_train)
```

```python
y_pred = grid.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, roc_auc_score

report = classification_report(y_test,y_pred)
confusionMatrix = confusion_matrix(y_test,y_pred)
roc_score = roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])

print(f"Classification Report:\n{report}\n")
print(f"Confusion Matrix:\n{confusionMatrix}\n")
print(f"ROC Score:\n{roc_score}")
```
