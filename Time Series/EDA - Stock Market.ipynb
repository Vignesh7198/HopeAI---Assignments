```python
def freqTable(columnName,stk_data):
    import pandas as pd
    frq = pd.DataFrame()
    frq["Unique_values"] = stk_data[columnName].value_counts().index
    frq["Frequency"] = stk_data[columnName].value_counts().values
    frq["Relative_fre"] = stk_data[columnName].value_counts().values/len(stk_data[columnName])*100
    frq["Cumulative"] = frq["Relative_fre"].cumsum()
    return frq
    
```

```python
freqTable("Low",stk_data).sort_values(by="Frequency",ascending=False)
```

```python
freqTable("High",stk_data).sort_values(by="Frequency",ascending=False)
```
